```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('stemmed_nouns_not_separated.csv')
```

```python
data.drop(data.columns[0],axis=1,inplace=True)
```

```python
data.iloc[0].loc['ingredients']
```

```python
data['ingredients'] =data['ingredients'].apply(lambda x : x.replace('[',''))
```

```python
data['ingredients']
```

```python
data
```

```python
cleaned_cuisine = data['ingredients'].str.split(',', expand=True).stack()
ing_binary =pd.get_dummies(cleaned_cuisine, prefix='in_').groupby(level=0).sum()
ing_binary.head(5)
```

```python
ing_binary
```

```python
X=ing_binary
y=data['cuisine']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=0)
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
labels = y.unique()
```

```python
labels
```

```python
confusion_matrix(y_test, y_pred,labels)
```

```python
cm = confusion_matrix(y_test, y_pred,labels)
print(cm)
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
plt.title('Confusion matrix of the classifier')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
